# Keras Introduction
Criando uma primeira rede utilizando *Keras*. Será utilizado **TensorFlow** como *backend*.

## Setup
Carregando as bibliotecas e fazendo o set de um *seed* (sempre bom setar um *seed* quando trabalhando com processos estocásticos, para sempre termos o mesmo resultado quando for rodado novamente).

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np

np.random.seed(7)
```

## Dados
Carregando os dados e separando: existem 8 colunas como *input* e a última é o *output*, onde o *output* é 1 ou 0, dizendo se a pessoa tem diabetes ou não.

```python
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")

# Separando a entrada X da saída Y
X = dataset[:, :8]
Y = dataset[:, 8]
```

## Definindo o modelo
Após carregar os dados e as bibliotecas necessárias é preciso definir o modelo que será utilizado para fazer a predição. 

Os modelos são definidos como uma sequencia de layers. É criado um modo *Sequential* e então é adicionado layers, um de cada vez, até que estejamos felizes com nossa rede. 

É necessário verificar que o *input layer* tenha o número correto de *inputs*. Isso pode ser especificado quando criando o *input layer* ao atribuir **8** ao argumento **input_dim**, já que temos 8 variáveis de entrada.

Neste exemplo será utilizado uma *Fully-Connected Network* (FC) com três layers. As FC's são definidas pela classe **Dense** e podemos especificar o número de neurônios na camada como primeiro argumento, o método de inicialização como segundo argumento por **init** e a função de ativação como terceiro argumento e **activation**. 

Nesse caso, iremos inicializar os *weights* da rede com um pequeno número aleatório gerado de uma distribuição uniforme (**uniform**), nesse caso entre 0 e 0.05, pois essa é a inicialização *default* de *weights* no *Keras* (por isso não será especificada no código abaixo). Outra alternativa seria '**normal**' para pequenos números gerados de uma distribuição *gaussiana*.

Será utilizado funções de ativação **ReLU** nas primeiras duas camadas e **sigmoid** na camada de saída.

```python
model = Sequential()
model.add(Dense(12, input_dim=8))
model.add(Activation('relu'))
model.add(Dense(8))
model.add(Activation('relu'))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

## Compilando o modelo
Agora que o modelo está definido, só falta compilar. O modelo compilado irá usar o backend **TensorFlow** para gerar o grafo computacional e treinar o modelo.

Será utilizado como *loss* a **binary_crossentropy**, e como método de atualização do gradiente **adam**. 

```python
# Compilando o modelo
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Treinando o algoritmo
Depois de definido e compilado, o treinamento do algoritmo é feito pelo método *fit()*, que receberá como parâmetros o número de *epochs* que deseja ser treinado, o tamanho do *batch_size*

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
```

## Avaliando o modelo
A avaliação do modelo pode ser feita pelo método *evaluate*

```python
model.evaluate(X, Y, verbose=0)
```

## Fazendo predições
Para fazer predições, basta utilizar a função *predict*.

```python
predictions = model.predict(X)
corr_predictions = [round(x[0]) for x in predictions]
correct = ['Diabetes' if x == 1 else 'Saudável' for x in corr_predictions]
for pacient in correct:
    print(pacient)
```
